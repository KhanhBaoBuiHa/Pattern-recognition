```python
from google.colab import files
uploaded = files.upload()   # Chọn file mpg.csv từ máy tính

# Đọc file CSV bằng pandas
import pandas as pd
df = pd.read_csv('mpg.csv')

print("5 dòng đầu tiên của dữ liệu")
print(df.head(), "\n")

# Kiểm tra giá trị bị thiếu
print("Kiểm tra giá trị bị thiếu")
missing = df.isnull().sum()
print(missing)

# Nếu có giá trị bị thiếu, ta xóa những dòng đó
df_clean = df.dropna()
print(f"\nSố dòng sau khi xóa giá trị thiếu: {len(df_clean)}")

# Kiểm tra dữ liệu trùng lặp
duplicates = df_clean.duplicated().sum()
print(f"\nSố dòng trùng lặp: {duplicates}")

# Xóa dòng trùng lặp (nếu có)
df_clean = df_clean.drop_duplicates()
print(f"Số dòng sau khi xóa trùng lặp: {len(df_clean)}")

# Thông tin tổng quan sau khi làm sạch
print("\nThông tin dữ liệu sau khi làm sạch")
print(df_clean.info())

# Lưu dữ liệu đã làm sạch để dùng cho câu b
df_clean.to_csv("mpg_clean.csv", index=False)
print("\nDữ liệu sạch đã lưu thành công vào file 'mpg_clean.csv'")
```

```python
import pandas as pd

# Đọc lại dữ liệu sạch
df = pd.read_csv('mpg_clean.csv')

# Tính giá trị trung bình và trung vị của 'cylinders', 'horsepower', 'weight'
mean_vals = df[['cylinders', 'horsepower', 'weight']].mean()
median_vals = df[['cylinders', 'horsepower', 'weight']].median()

print("Trung bình và Trung vị")
print("Giá trị trung bình:\n", mean_vals, "\n")
print("Giá trị trung vị:\n", median_vals, "\n")

# Miền giá trị (min, max) và phân vị (Q1, Q3) của 'mpg'
mpg_min = df['mpg'].min()
mpg_max = df['mpg'].max()
q1 = df['mpg'].quantile(0.25)
q3 = df['mpg'].quantile(0.75)

print("Thống kê 'mpg'")
print(f"Min: {mpg_min}, Max: {mpg_max}")
print(f"Q1: {q1}, Q3: {q3}\n")

# Tìm xe có mức tiêu hao nhiên liệu đặc biệt cao hoặc thấp
print("Xe có mức tiêu hao nhiên liệu thấp nhất:")
print(df[df['mpg'] == mpg_min][['name', 'mpg', 'origin']])
print("\nXe có mức tiêu hao nhiên liệu cao nhất:")
print(df[df['mpg'] == mpg_max][['name', 'mpg', 'origin']])
```

```python
# Trung bình mpg của từng nhóm 'origin'
mpg_by_origin = df.groupby('origin')['mpg'].mean()
print("Trung bình mpg theo khu vực")
print(mpg_by_origin)
```

```python
import matplotlib.pyplot as plt

# 5Histogram phân bố trọng lượng (weight)
plt.figure(figsize=(8,5))
plt.hist(df['weight'], bins=20)
plt.title("Phân bố trọng lượng xe (weight)")
plt.xlabel("Trọng lượng (pound)")
plt.ylabel("Số lượng xe")
plt.grid(True)
plt.show()

# Boxplot so sánh công suất (horsepower) giữa thập niên 70 và 80
# Tạo cột 'decade' = 70 hoặc 80 dựa theo 'model_year'
df['decade'] = df['model_year'].apply(lambda x: 70 if x < 80 else 80)

plt.figure(figsize=(6,5))
df.boxplot(column='horsepower', by='decade')
plt.title("So sánh công suất giữa thập niên 70 và 80")
plt.suptitle("")
plt.xlabel("Thập niên")
plt.ylabel("Công suất (horsepower)")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))

# Tạo 4 biểu đồ scatter
features = ['cylinders', 'horsepower', 'acceleration', 'model_year']

for i, col in enumerate(features):
    plt.subplot(2,2,i+1)
    sns.scatterplot(data=df, x=col, y='mpg')
    plt.title(f'mpg vs {col}')

plt.tight_layout()
plt.show()
```
