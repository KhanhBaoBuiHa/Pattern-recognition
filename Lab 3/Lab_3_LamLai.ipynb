```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson, binom, norm
```

```python
#Đọc dữ liệu
mpg_df = pd.read_csv(r'D:\Zalo\mpg.csv')
```

```python
# CÂU A

print("\nKiem tra du lieu thieu & trung lap")
missing_values = mpg_df.isnull().sum()
print("So luong gia tri bi thieu: ")
print(missing_values)

#Xoa nhung dong bi thieu
df_clean = mpg_df.dropna()
print(f"So dong sau khi xoa gia tri bi thieu: {len(df_clean)}")

#Kiem tra du lieu trung lap
duplicates = df_clean.duplicated().sum() 
print(f"So dong trung lap:  {duplicates}")

#Xoa nhung dong trung lap
df_clean = df_clean.drop_duplicates()
print(f"So dong sau khi xoa trung lap: {len(df_clean)}")
```

```python
# CÂU B
mean_values = mpg_df[['cylinders','horsepower','weight']].mean() #trung bình
median_values = mpg_df[['cylinders','horsepower','weight']].median() #trung vị
print("Gia tri trung binh:\n",mean_values, "\n")
print("Gia tri trung vi:\n",median_values, "\n")

mpg_range = (mpg_df['mpg'].min(), mpg_df['mpg'].max())
mpg_q1 = mpg_df['mpg'].quantile(0.25)
mpg_q3 = mpg_df['mpg'].quantile(0.75)

print("Thong ke\n")
print(mpg_range)
print(f"Phan vi Q1, Q3 cua 'mpg':({mpg_q1}, {mpg_q3})\n")
```

```python
# CÂU C
print("\n Tieu hao nhien lieu trung binh theo xuat xu")
mpg_by_origin = mpg_df.groupby('origin')['mpg'].mean()
print(mpg_by_origin)
print("Khu vuc it tieu hao nhien lieu nhat :", mpg_by_origin.idxmin())
```

```python
# CÂU D
print("\nBiểu đồ phân phối trọng lượng & công suất")
plt.figure(figsize=(12,5))
sns.histplot(mpg_df['weight'], bins=20, kde=True)
plt.title("Phân bố trọng lượng xe")
plt.xlabel("Trọng lượng (pound)")
plt.ylabel("Số lượng xe")
plt.show

plt.figure(figsize=(12,5))
sns.boxplot(x=mpg_df['model_year'] //10*10, y=mpg_df['horsepower'])
plt.title("So sánh công suất giữa các xe trong thập niên 70 và 80")
plt.xlabel("Thap nien san xuat")
plt.ylabel("Cong suat")
plt.show()
```
